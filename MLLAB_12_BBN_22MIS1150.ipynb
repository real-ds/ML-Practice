```python
import pandas as pd

# Load the dataset (ensure path is correct)
data = pd.read_excel("/content/drive/MyDrive/ML LAB DATASETs/data_bn.xlsx", engine="openpyxl")

# Display initial dataset
print(data.head())

# Encode categorical variables using Label Encoding
for col in data.columns:
    data[col] = data[col].astype("category").cat.codes  # Convert categorical to numeric

# Display processed dataset
print(data.head())
```

```python
import bnlearn as bn

# Define the Bayesian Network structure
edges = [
    ('sex', 'smoke'),  # Sex may influence smoking habits
    ('age', 'smoke'),  # Age may influence smoking habits
    ('smoke', 'asthma'),  # Smoking can cause asthma
    ('allergy', 'asthma'),  # Allergy may contribute to asthma
    ('urbanization', 'asthma'),  # Air quality in urban areas may impact asthma
    ('sedentary', 'asthma'),  # Sedentary lifestyle may contribute to asthma
    ('education', 'smoke'),  # Education level may influence smoking habits
]

# Create the Bayesian Network
model = bn.make_DAG(edges)

# Plot the structure
bn.plot(model)
```

```python
# Train the model with the dataset
model_trained = bn.parameter_learning.fit(model, data)

# Print learned CPDs (Conditional Probability Distributions)
bn.print_CPD(model_trained)
```

```python
# Define evidence (observed values)
evidence = {'smoke': 1}

# Perform inference
inference_result = bn.inference.fit(model_trained, variables=['asthma'], evidence=evidence)

print(inference_result)
```

```python
# Define evidence (observed values)
evidence = {'smoke': 0}

# Perform inference
inference_result = bn.inference.fit(model_trained, variables=['asthma'], evidence=evidence)

print(inference_result)
```
