```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

def evaluate_model(X, y, dataset_name):
    """
    Trains NaÃ¯ve Bayes and evaluates it on different train-test splits
    """
    for split in [0.3, 0.2, 0.1]:  # Corresponds to 70-30, 80-20, 90-10 splits
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=split, random_state=42)

        # Train model
        model = GaussianNB()
        model.fit(X_train, y_train)

        # Predictions
        y_pred = model.predict(X_test)

        # Evaluation
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, average="weighted")
        recall = recall_score(y_test, y_pred, average="weighted")
        f1 = f1_score(y_test, y_pred, average="weighted")
        conf_matrix = confusion_matrix(y_test, y_pred)

        # Print Results
        print(f"\nðŸ“Œ **{dataset_name} Dataset ({100-int(split*100)}-{int(split*100)}) Split Results**")
        print(f"Accuracy: {accuracy:.4f}")
        print(f"Precision: {precision:.4f}")
        print(f"Recall: {recall:.4f}")
        print(f"F1 Score: {f1:.4f}")
        print("Confusion Matrix:\n", conf_matrix)

        # Plot Confusion Matrix
        plt.figure(figsize=(5,4))
        sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
        plt.xlabel("Predicted")
        plt.ylabel("Actual")
        plt.title(f"Confusion Matrix - {dataset_name} ({100-int(split*100)}-{int(split*100)})")
        plt.show()
```

```python
df1 = pd.read_csv("/content/drive/MyDrive/ML LAB DATASETs/Iris.csv")  # Replace with actual file path
df1.drop(columns=['Id'], inplace=True)  # Remove unnecessary column

# Split features & target
X1 = df1.drop(columns=['Species'])
y1 = df1['Species']

# Evaluate on different splits
evaluate_model(X1, y1, "Iris")
```

```python
df2 = pd.read_csv("/content/drive/MyDrive/ML LAB DATASETs/KNNAlgorithmDataset.csv")  # Replace with actual file path
df2.drop(columns=['id', 'Unnamed: 32'], inplace=True)  # Remove unnecessary columns

# Encode Target Variable (M â†’ 1, B â†’ 0)
X2 = df2.drop(columns=['diagnosis'])
y2 = df2['diagnosis'].map({'M': 1, 'B': 0})

# Evaluate on different splits
evaluate_model(X2, y2, "Breast Cancer")
```

```python
df3 = pd.read_csv("/content/drive/MyDrive/ML LAB DATASETs/Naive-Bayes-Classification-Data.csv")  # Replace with actual file path

# Split features & target
X3 = df3.drop(columns=['diabetes'])
y3 = df3['diabetes']

# Evaluate on different splits
evaluate_model(X3, y3, "Diabetes")
```
