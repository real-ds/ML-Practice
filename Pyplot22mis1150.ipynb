```python
import numpy as np
import matplotlib.pyplot as plt

List1 = np.arange(0, 10, 0.1)
List2 = np.sin(List1)
List3 = np.cos(List1)

plt.plot(List1, List2, label='sin') # Plotting the sine curve
plt.plot(List1, List3, label='cos') # Plotting the cosine curve

plt.title("Sin and Cos Curve on List1")
plt.xlabel("0 to 10 (0.1)")
plt.ylabel("-1 to 1")
plt.legend()  # Adding a legend to differentiate the curves
plt.show()
```

```python
x=np.arange(0,100,2)
y=np.arange(50,-50,-2)
y1=np.arange(100,0,-2)
plt.title("With labels and different colors")
plt.xlabel("Range 0 to 50")
plt.ylabel("Range 50 to -50 ")
plt.scatter(x,y,label="one",color='green')
plt.scatter(x,y1,label="one",color='red')
plt.legend(['A','B'])
plt.show()
```

```python
x=np.arange(0,100,2)
y=np.arange(50,-50,-2)
y1=np.arange(100,0,-2)
plt.figure(figsize=(5,5))
plt.title("With Gridlines and sized")
plt.xlabel("Range 0 to 50")
plt.ylabel("Range 50 to -50 ")

plt.grid(True)
plt.scatter(x,y,label="one",color='green')
plt.scatter(x,y1,label="one",color='red')
plt.legend(['A','B'])
plt.show()
```

```python
figure=plt.figure()
axes1=figure.add_subplot(1,1,1)
axes2=axes1.twinx()

List1=np.linspace(0.,10.,100)
List2=[x**2 for x in List1]
List3=[x**3 for x in List1]

axes1.plot(List1,List2,'b-')
axes2.plot(List1,List3,'r-')

axes1.set_ylabel('x power 2',color="red")
axes2.set_ylabel('x power 3',color="blue")
axes1.set_xlabel('x')
```

```python
x=np.linspace(-3.14,3.14,100)
y=np.sin(x)
z=np.cos(x)
#plotting Curves
plt.plot(x,y,linewidth=5,c='#FA4500',linestyle='--')
plt.plot(x,z,c='b',marker='|',markersize=10)
plt.show()
```
