Implementation of Non linear SVM Algorithm on Kernel = RBF

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Load dataset
dataset = "/content/drive/MyDrive/ML LAB DATASETs/KNNAlgorithmDataset.csv"  # Update with your dataset path
df = pd.read_csv(dataset)

# Drop unnecessary columns
df = df.drop(columns=["id", "Unnamed: 32"], errors="ignore")

# Convert diagnosis column (M = 1, B = 0)
df["diagnosis"] = df["diagnosis"].map({"M": 1, "B": 0})

# Separate features and target
X = df.drop(columns=["diagnosis"])
y = df["diagnosis"]

# Function to apply Non-Linear SVM with different train-test splits
def apply_nonlinear_svm(test_size):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42, stratify=y)

    # Feature Scaling (Standardization)
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    # Train Non-Linear SVM model with RBF Kernel
    svm_model = SVC(kernel="rbf", random_state=42, gamma="scale", C=1.0)  # You can tune gamma and C
    svm_model.fit(X_train, y_train)

    # Predictions
    y_pred = svm_model.predict(X_test)

    # Model Evaluation
    accuracy = accuracy_score(y_test, y_pred)
    print(f"\nTrain-Test Split: {int((1-test_size)*100)}-{int(test_size*100)}")
    print(f"Accuracy: {accuracy:.4f}")
    print("Classification Report:\n", classification_report(y_test, y_pred))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Benign", "Malignant"], yticklabels=["Benign", "Malignant"])
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title(f"Confusion Matrix ({int((1-test_size)*100)}-{int(test_size*100)})")
    plt.show()

# Apply for different train-test splits
for split in [0.3, 0.2, 0.1]:
    apply_nonlinear_svm(split)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Load dataset
dataset = "/content/drive/MyDrive/ML LAB DATASETs/KNNAlgorithmDataset.csv"  # Update with your dataset path
df = pd.read_csv(dataset)

# Drop unnecessary columns
df = df.drop(columns=["id", "Unnamed: 32"], errors="ignore")

# Convert diagnosis column (M = 1, B = 0)
df["diagnosis"] = df["diagnosis"].map({"M": 1, "B": 0})

# Separate features and target
X = df.drop(columns=["diagnosis"])
y = df["diagnosis"]

# Kernels and Splits to test
kernels = ["rbf", "poly", "sigmoid"]
splits = [0.3, 0.2, 0.1]  # 70-30, 80-20, 90-10

# Store metrics for comparison
results = []

# Function to apply Non-Linear SVM and collect metrics
def apply_nonlinear_svm(test_size, kernel):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42, stratify=y)

    # Feature Scaling (Standardization)
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    # Train Non-Linear SVM model
    svm_model = SVC(kernel=kernel, random_state=42, gamma="scale", C=1.0)
    svm_model.fit(X_train, y_train)

    # Predictions
    y_pred = svm_model.predict(X_test)

    # Model Evaluation
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, output_dict=True)  # Dictionary format

    # Store results
    results.append({
        "Kernel": kernel.upper(),
        "Split": f"{int((1-test_size)*100)}-{int(test_size*100)}",
        "Accuracy": accuracy,
        "Precision": report["weighted avg"]["precision"],
        "Recall": report["weighted avg"]["recall"],
        "F1-Score": report["weighted avg"]["f1-score"]
    })

# Apply for each kernel and split
for kernel in kernels:
    for split in splits:
        apply_nonlinear_svm(split, kernel)

# Convert results to DataFrame
df_results = pd.DataFrame(results)

# Plot Bar Graphs for Accuracy, Precision, Recall, and F1-score
metrics = ["Accuracy", "Precision", "Recall", "F1-Score"]
plt.figure(figsize=(14, 8))

for i, metric in enumerate(metrics, 1):
    plt.subplot(2, 2, i)
    sns.barplot(data=df_results, x="Split", y=metric, hue="Kernel")
    plt.title(f"Comparison of {metric} for Different Kernels & Splits")
    plt.xlabel("Train-Test Split")
    plt.ylabel(metric)
    plt.legend(title="Kernel")

plt.tight_layout()
plt.show()
```
